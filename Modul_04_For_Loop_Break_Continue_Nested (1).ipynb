## 1. Recap Kilat: While Loop & Kapan Memilihnya

Sebelum masuk ke `for`, mari ingat kembali `while`. `while` loop terus berjalan **selama kondisi masih `True`**, dan kita wajib mengubah kondisi itu di dalam loop agar tidak terjadi *infinite loop*.

> **Kapan pakai `while` vs `for`?**
> | | `while` | `for` |
> |---|---|---|
> | Kapan pakai | Jumlah perulangan **tidak diketahui** | Jumlah perulangan **sudah diketahui** |
> | Contoh | Validasi input user | Cetak angka 1–10 |
> | Kondisi berhenti | Kondisi menjadi `False` | Iterasi habis |

````python
# Recap: while loop - validasi input angka 1-10
num = int(input("Masukkan angka antara 1-10: "))

while num < 1 or num > 10:
    print(f"{num} tidak valid! Angka harus antara 1 dan 10.")
    num = int(input("Masukkan angka antara 1-10: "))

print(f"Bagus! Kamu memilih angka {num}")
````

## 2. Teori: For Loop & `range()`

`for` loop digunakan untuk **mengulang sebanyak jumlah yang sudah ditentukan**. Biasanya dikombinasikan dengan fungsi `range()` yang menghasilkan urutan angka.

> **Analogi:** Kalau `while` itu seperti menjaga toko *"sampai ada pelanggan datang"* (tidak tahu kapan berhenti), maka `for` itu seperti mengantarkan koran ke **10 rumah** — kamu tahu persis akan berhenti setelah rumah ke-10.

**Sintaks dasar:**
```python
for variabel in range(start, stop):
    # kode yang diulang
```

> **Common Mistake:** Sama seperti slicing, nilai `stop` pada `range()` bersifat *exclusive* (tidak termasuk). Jadi `range(1, 11)` menghasilkan angka 1 sampai **10**, bukan 11!

> **Tips:** Nama variabel `x` bisa diganti apapun. Pilih nama yang bermakna, misalnya `for hari in range(1, 8)` lebih jelas daripada `for x in range(1, 8)`.

````python
# Demo 1a: Menghitung 1 sampai 10
print("Hitung maju:")
for x in range(1, 11):
    print(x, end=" ")  # end=" " agar semua angka tampil dalam satu baris

# Demo 1b: Nama variabel yang lebih bermakna
print("\n\nDaftar hari dalam seminggu:")
for hari in range(1, 8):
    print(f"Hari ke-{hari}")
````

````python
i = 1
while i <= 10:
    print(i, end=" ")
    i += 1  # Increment i by 1
````

## 3. Variasi `range()`: reversed & step

`range()` punya dua variasi yang membuat `for` loop jauh lebih fleksibel.

### A. `reversed()` — Urutan Terbalik

Fungsi `reversed()` membungkus `range()` agar iterasinya berjalan dari belakang ke depan.

````python
# Demo 2: Hitung mundur dari 10 ke 1
print("Hitung mundur:")
for x in reversed(range(1, 11)):
    print(x, end=" ")

print("\nSelamat Tahun Baru! 🎉")
````

### B. Step (Langkah) — Lompatan Antar Angka

`step` adalah argumen **ketiga** di `range()` — menentukan seberapa besar lompatan setiap iterasi.

**Sintaks lengkap:** `range(start, stop, step)`

> **Pertanyaan Refleksi:** Bagaimana cara mencetak angka **genap** dari 2 sampai 20 menggunakan `range()` dengan step? Coba tulis sendiri sebelum melihat demo di bawah!

````python
# Demo 3a: Cetak angka GANJIL dari 1-10 (mulai dari 1, lompat 2)
print("Angka ganjil 1-10:")
for x in range(1, 11, 2):
    print(x, end=" ")

# Demo 3b: Cetak angka GENAP dari 2-20 (mulai dari 2, lompat 2)
print("\n\nAngka genap 2-20:")
for x in range(2, 21, 2):
    print(x, end=" ")

# Demo 3c: Step negatif (hitung mundur langsung dengan range)
print("\n\nHitung mundur dari 10 ke 1 dengan step negatif:")
for x in range(10, 0, -1):  # start=10, stop=0 (exclusive), step=-1
    print(x, end=" ")
````

### Exercise 1: Countdown Roket 🚀

````python
# EXERCISE 1: Countdown Roket
# Cetak hitungan mundur dari 10 ke 1 menggunakan step negatif,
# lalu cetak "🚀 Roket meluncur!"

for x in range(10, 0, -1):
    print(x)
print("🚀 Roket meluncur!")
````

## 4. Teori: For Loop pada String

`for` loop tidak hanya bisa mengiterasi angka — kita juga bisa mengiterasi setiap **karakter** dalam sebuah string, satu per satu, dari kiri ke kanan!

> **Analogi:** Bayangkan kamu membaca sebuah kata dengan menunjuk menggunakan jari — jari berpindah satu huruf per huruf, dari kiri ke kanan. Itulah yang dilakukan `for` loop pada string.

Ini sangat berguna untuk: menganalisis teks, memvalidasi karakter, menghitung sesuatu dalam string, dll.

````python
# Demo 4a: Iterasi setiap karakter dalam string
credit_card = "1234-5678-9012-3456"

print("Karakter satu per satu:")
for x in credit_card:
    print(x)

# Demo 4b: Hitung jumlah tanda '-' dalam nomor kartu
jumlah_strip = 0
for x in credit_card:
    if x == "-":
        jumlah_strip += 1  # jumlah_strip = jumlah_strip + 1

print(f"\nJumlah tanda '-': {jumlah_strip}")
````

### Exercise 2: Hitung Huruf Vokal

````python
# EXERCISE 2: Hitung jumlah huruf vokal dalam sebuah kata
# Huruf vokal: a, i, u, e, o (gunakan .lower() agar tidak case-sensitive)

kata = input("Masukkan sebuah kata: ")
jumlah_vokal = 0

for huruf in kata.upper():
    if huruf in "AIUEO":    # cek apakah huruf ada dalam string vokal
        jumlah_vokal += 1

print(f"Kata '{kata}' memiliki {jumlah_vokal} huruf vokal")
````

## 5. Teori: `continue` & `break`

Dua keyword ini digunakan untuk **mengontrol alur** di dalam loop:

| Keyword | Fungsi | Analogi |
|---------|--------|---------|
| `continue` | **Lewati** iterasi saat ini, lanjut ke iterasi berikutnya | Melewati telur retak, lanjut ke telur berikutnya |
| `break` | **Hentikan** loop sepenuhnya, langsung keluar | Menemukan telur busuk, langsung berhenti |

> **Tips penting:** `continue` dan `break` hanya mempengaruhi loop **terdalam** tempatnya berada. Jika ada nested loop, `break` di inner loop hanya menghentikan inner loop, bukan outer loop.

````python
# Demo 5a: continue — SKIP angka 13 saja, lanjut ke 14
print("Angka 1-20 (13 di-skip):")
for x in range(1, 21):
    if x == 13:
        continue   # lewati iterasi ini, kembali ke atas loop
    print(x, end=" ")
````

````python
# Demo 5b: break — BERHENTI total saat ketemu 13
print("Angka 1-20 (berhenti di 13):")
for x in range(1, 21):
    if x == 13:
        break      # keluar dari loop sepenuhnya
    print(x, end=" ")

print("\nLoop sudah selesai")
````

> **Perbandingan visual:**
> - `continue` → `[1, 2, 3, ..., 12, _, 14, 15, ..., 20]` (13 kosong, lanjut)
> - `break`    → `[1, 2, 3, ..., 12, STOP]` (berhenti di 12, 13 tidak dicetak)

### Exercise 3: Validator Karakter Username

````python
# EXERCISE 3: Cek apakah username mengandung spasi atau karakter '@'
# - Jika ya  → cetak karakter bermasalah + hentikan dengan break
# - Jika tidak ada → cetak "Username valid!"

username = input("Masukkan username: ")
valid = True

for char in username:
    if char == " " or char == "@":
        print(f"❌ Username tidak valid! Karakter '{char}' tidak diperbolehkan.")
        valid = False
        break  # tidak perlu cek karakter sisanya

if valid:
    print("✅ Username valid!")
````

## 6. Teori: Nested Loop (Perulangan Bersarang)

**Nested loop** adalah loop di dalam loop — ada **outer loop** (luar) dan **inner loop** (dalam). Cara kerjanya: setiap **satu kali** iterasi outer loop, inner loop **berjalan penuh dari awal sampai akhir**.

> **Analogi:** Bayangkan jarum jam:
> - **Outer loop** = jarum jam → bergerak 1 langkah setiap 60 menit
> - **Inner loop** = jarum menit → berputar penuh 60 kali sebelum jarum jam bergerak 1 langkah

```
Outer ke-1: inner berjalan penuh (1,2,3,...)
Outer ke-2: inner berjalan penuh lagi (1,2,3,...)
Outer ke-3: inner berjalan penuh lagi (1,2,3,...)
...
```

````python
# Demo 6a: Grid simbol — outer = baris, inner = kolom
rows   = int(input("Masukkan jumlah baris: "))
#3
columns = int(input("Masukkan jumlah kolom: "))
#3
symbol = input("Masukkan simbol: ")
#?

for x in range(rows):            # outer loop: mengontrol BARIS
    for y in range(columns):     # inner loop: mengontrol KOLOM
        print(symbol, end="")   # end="" agar tidak pindah baris
    print()                      # pindah baris setelah 1 baris selesai
    
# 3 baris
# 3 kolom

# baris 1 - ? ? ?
# baris 2 - ? ? ?
# baris 3 - ? ? ?
````

````python
# Demo 6b: Tabel perkalian sederhana (3x3)
# Ini contoh nyata nested loop yang sering dipakai!

print("Tabel Perkalian 1-5:")
print("-" * 30)

for baris in range(1, 6):          # outer: angka pengali kiri (1-5)
    for kolom in range(1, 6):      # inner: angka pengali kanan (1-5)
        hasil = baris * kolom
        print(f"{hasil:3}", end="")  # :3 agar setiap angka memakai 3 karakter
    print()                        # pindah baris
    
    # baris 1 - 5
    # column 1 - 5
    
    # 1 = 1 (1x1) 2 (1x2) 3 (1x3) 4 (1x4) 5 (1x5)
    # 2 = 2 (2x1) 4 (2x2) 6 (2x3) 8 (2x4) 10 (2x5)
    # 3 = 3 (3x1) 6 (3x2) 9 (3x3) 12 (3x4) 15 (3x5)
    # 4 = 4 (4x1) 8 (4x2) 12 (4x3) 16 (4x4) 20 (4x5)
    # 5 = 5 (5x1) 10 (5x2) 15 (5x3) 20 (5x4) 25 (5x5)
    
````

> **Common Mistake:** Jangan lupa `print()` kosong di akhir **outer loop** untuk pindah baris! Tanpa itu, semua output akan tercetak dalam satu baris panjang.

### Exercise 4: Pola Segitiga Siku-Siku

````python
# EXERCISE 4: Pola Segitiga Siku-Siku
# Kunci: jumlah kolom di setiap baris = nomor barisnya
# Baris 1 → 1 bintang
# Baris 2 → 2 bintang
# Baris 3 → 3 bintang, dst.

baris = int(input("Masukkan jumlah baris: "))

for x in range(1, baris + 1):   # x = 1, 2, 3, ..., baris
    for y in range(x):          # inner loop berjalan x kali
        print("*", end="")
    print()
    
# 5 
# (1,6)

# X = 1
#     y = 1

# *

# x = 2
#     y = 1 , 2
# **
````

## 7. Mini Project: Countdown Timer ⏱️

Kita gabungkan semua yang sudah dipelajari: `for` loop, `range()` dengan step negatif, format specifier dari Modul 3, dan modul `time` untuk membuat countdown timer dalam format `HH:MM:SS`!

**Logika konversi detik ke jam:menit:detik:**

| Variabel | Rumus | Contoh (input: 3723 detik) |
|----------|-------|----------------------------|
| `seconds` | `x % 60` | `3723 % 60 = 3` |
| `minutes` | `int(x / 60) % 60` | `int(3723/60) % 60 = 2` |
| `hours` | `int(x / 3600)` | `int(3723/3600) = 1` |

Format `:02` artinya: tampilkan minimal 2 digit, tambahkan nol di depan jika perlu (`5` → `05`).

````python
import time

my_time = int(input("Masukkan waktu dalam detik: "))

for x in range(my_time, 0, -1):
    seconds = x % 60
    minutes = int(x / 60) % 60
    hours   = int(x / 3600)

    # Tampilan format HH:MM:SS
    print(f"{hours:02}:{minutes:02}:{seconds:02}")

    # Pesan peringatan 10 detik terakhir
    if x == 10:
        print("⚠️  10 detik lagi!")

    time.sleep(1)   # jeda 1 detik sebelum iterasi berikutnya

print("⏰ WAKTU HABIS!")
````

> **Tantangan Tambahan:** Modifikasi countdown timer di atas agar:
> 1. Menampilkan pesan `"⚠️ 10 detik lagi!"` saat `x == 10` *(sudah ada di kode di atas!)*
> 2. Menampilkan `"Setengah jalan!"` tepat saat waktu tersisa setengah dari input awal
> 3. Countdown **berhenti** jika user mengetik sesuatu *(tantangan bonus tingkat lanjut!)*

## 8. Pengenalan Koleksi Data: List, Set, dan Tuple

Selama ini kita menyimpan data satu per satu dalam satu variabel. Bagaimana kalau kita punya **banyak data** yang saling berhubungan? Di sinilah tiga tipe koleksi data Python berperan!

| Tipe | Simbol | Ordered | Changeable | Duplikat | Kecepatan |
|------|--------|---------|------------|----------|-----------|
| **List** | `[ ]` | ✅ Ya | ✅ Ya | ✅ Boleh | Normal |
| **Set** | `{ }` | ❌ Tidak | ⚠️ Add/Remove OK | ❌ Tidak boleh | Cepat untuk cek keanggotaan |
| **Tuple** | `( )` | ✅ Ya | ❌ Tidak | ✅ Boleh | ✅ Paling cepat |

> **Analogi:**
> - **List** = keranjang belanja — bisa tambah/hapus item, urutan penting, boleh ada duplikat
> - **Set** = kotak stempel kepesertaan — hanya nama unik, urutan tidak penting
> - **Tuple** = koordinat GPS `(lat, long)` — sudah pasti, tidak boleh diubah

### A. List `[ ]` — Koleksi Berurutan yang Fleksibel

````python
# List: ordered, changeable, duplicates OK
buah = ["apel", "mangga", "jeruk", "apel"]  # duplikat 'apel' boleh!

print("List awal    :", buah)
print("Index ke-0   :", buah[0])     # akses elemen pertama
print("Index ke-1   :", buah[1])
print("Index terakhir:", buah[-1])   # index negatif dari belakang
print("Panjang list :", len(buah))

# Mengubah elemen (ini tidak bisa dilakukan di tuple!)
buah[1] = "anggur"
print("\nSetelah buah[1] diganti:", buah)

# Menambah elemen di akhir
buah.append("pisang")
print("Setelah append 'pisang':", buah)

# Menghapus elemen berdasarkan nilai
buah.remove("jeruk")
print("Setelah remove 'jeruk' :", buah)
````

````python
# For loop pada List: sangat natural dan sering dipakai!
belanjaan = ["kopi", "roti", "susu", "telur", "gula"]

print("🛒 Daftar Belanjaan:")
for item in belanjaan:
    print(f"  - {item}")

print(f"\nTotal: {len(belanjaan)} item")
````

### B. Set `{ }` — Koleksi Unik Tanpa Duplikat

````python
# Set: unordered, no duplicates, bisa add/remove
# Perhatikan: ada banyak duplikat saat membuat set di bawah
pengunjung = {"Budi", "Ani", "Caca", "Budi", "Ani", "Dedi"}

print("Set pengunjung:", pengunjung)
# Perhatikan: duplikat otomatis dihapus! Urutan bisa berbeda setiap run.

# Menambah anggota baru
pengunjung.add("Eko")
print("Setelah add 'Eko'  :", pengunjung)

# Menghapus anggota
pengunjung.remove("Caca")
print("Setelah remove 'Caca':", pengunjung)

# Keunggulan set: cek keanggotaan sangat cepat!
print("\nApakah 'Budi' hadir?:", "Budi" in pengunjung)
print("Apakah 'Caca' hadir?:", "Caca" in pengunjung)
````

````python
# DEMO: Set otomatis menghapus duplikat — berguna untuk data cleaning!
nilai_raw = [85, 92, 78, 92, 85, 67, 100, 78, 92]
print("Data asli (ada duplikat):", nilai_raw)

# Konversi ke set: duplikat hilang seketika
nilai_unik = set(nilai_raw)
print("Nilai unik (pakai set)  :", nilai_unik)
print(f"Dari {len(nilai_raw)} data menjadi {len(nilai_unik)} data unik")
````

### C. Tuple `( )` — Koleksi Tetap yang Tidak Bisa Diubah

````python
# Tuple: ordered, UNCHANGEABLE, lebih cepat dari list

# Contoh penggunaan tuple: data yang tidak boleh berubah
koordinat = (-6.2088, 106.8456)   # latitude, longitude Jakarta
warna_merah = (255, 0, 0)         # nilai RGB
hari_kerja  = ("Senin", "Selasa", "Rabu", "Kamis", "Jumat")

print("Koordinat Jakarta :", koordinat)
print("Latitude          :", koordinat[0])   # akses index seperti list
print("Longitude         :", koordinat[1])

print("\nWarna Merah (RGB) :", warna_merah)

print("\nHari kerja:")
for hari in hari_kerja:
    print(f"  - {hari}")

# Baris berikut akan menghasilkan ERROR karena tuple tidak bisa diubah
# Uncomment untuk melihat error-nya:
# koordinat[0] = 0  # TypeError: 'tuple' object does not support item assignment
````

### D. For Loop pada Semua Tipe Koleksi

````python
# For loop bekerja pada List, Set, dan Tuple!

my_list  = [10, 20, 30, 40, 50]
my_set   = {100, 200, 300, 400, 500}
my_tuple = ("a", "b", "c", "d", "e")

print("Iterasi List (urutan terjaga):")
for item in my_list:
    print(item, end=" ")

print("\n\nIterasi Set (urutan TIDAK dijamin sama setiap run):")
for item in my_set:
    print(item, end=" ")

print("\n\nIterasi Tuple (urutan terjaga):")
for item in my_tuple:
    print(item, end=" ")
````

### Exercise 5: Cari Nilai Tertinggi dalam List

````python
# EXERCISE 5: Cari nilai tertinggi dari list nilai TANPA menggunakan fungsi max()
# Logika: anggap elemen pertama sebagai tertinggi sementara,
# lalu bandingkan satu per satu dengan elemen berikutnya

nilai = [78, 92, 85, 67, 95, 88, 71, 99, 60]
print("Data nilai:", nilai)

tertinggi = nilai[0]    # asumsikan index 0 sebagai nilai tertinggi awal

for n in nilai:
    if n > tertinggi:
        tertinggi = n   # perbarui jika ketemu yang lebih besar

print(f"Nilai tertinggi: {tertinggi}")
````

> **Ringkasan: Kapan Pakai Apa?**
> - **List** → saat kamu perlu koleksi yang bisa diubah-ubah dan urutannya penting
> - **Set** → saat kamu butuh data **unik** atau sering melakukan **cek keanggotaan** (`in`)
> - **Tuple** → saat data **tidak boleh berubah** (koordinat, konfigurasi, warna) atau butuh performa lebih cepat
