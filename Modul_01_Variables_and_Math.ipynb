## 1. Teori: Variabel dan Tipe Data
**Variabel** adalah sebuah wadah atau kontainer untuk menyimpan suatu nilai. Sebuah variabel akan berperilaku layaknya nilai yang ada di dalamnya.

Di Python, ada 4 tipe data dasar yang paling sering digunakan:
* **String (`str`)**: Teks atau karakter (diapit tanda kutip).
* **Integer (`int`)**: Bilangan bulat.
* **Float (`float`)**: Bilangan desimal.
* **Boolean (`bool`)**: Nilai kebenaran (`True` atau `False`).

```python
# Strings
first_name = "Bro"
food = "pizza"

# Integers
age = 25
quantity = 3

# Float
gpa = 3.2

# Boolean
is_student = True


# Menampilkan output
print(f"Hello {first_name}")
print(f"You are {age} years old")
print(f"Your gpa is: {gpa}")

if is_student:
    print("You are a student")
else:
    print("You are NOT student")
```

## 2. Teori: Type Casting
**Type Casting** adalah proses mengonversi sebuah nilai dari satu tipe data ke tipe data lainnya (misalnya dari String ke Integer)[cite: 1]. 

Hal ini sangat berguna, terutama nanti ketika kita menerima input dari pengguna (yang secara default selalu berupa String) dan ingin melakukan operasi matematika pada input tersebut. Kita bisa mengecek tipe data saat ini menggunakan fungsi `type()`[cite: 1].

```python

# Nilai awal
name = "Bro"
age = 21
gpa = 1.9
student = True

# Mengecek tipe data awal
print("--- Tipe Data Awal ---")
print(type(name))
print(type(age))
print(type(gpa))
print(type(student)) 

print("\n--- Setelah Type Casting ---")
# Mengubah integer menjadi float
age = float(age)

print(f"Age (float): {age}")

# Mengubah float menjadi integer (membulatkan ke bawah / membuang desimal)
gpa = int(gpa)
print(f"GPA (int): {gpa}")

# Mengubah boolean menjadi string
student = str(student)
print(f"Student (str): {student} -> tipenya kini {type(student)}")

# Mengubah string menjadi boolean (String tidak kosong bernilai True)
name = bool(name)
print(f"Name (bool): {name}")

ewa = 1

print(ewa)

ewa = float(ewa)

print(f"ewa (float): {ewa} -> tipenya kini {type(ewa)}")
```


## 3. Matematika Dasar di Python

### A. Operator Aritmatika & Augmented Assignment
Python menggunakan operator standar untuk matematika (`+`, `-`, `*`, `/`, `**` untuk pangkat, dan `%` untuk modulus/sisa bagi). 
Untuk mempersingkat penulisan saat mengubah nilai variabel itu sendiri, kita bisa menggunakan *augmented assignment operators* seperti `+=` atau `*=`.

```python
#biner

a = 10
b = 3

c = a + b

print (c)

# ============

x = 12
y = 5

print(x + y)
#16
print(x - y)
#6
print(x * y)
#55
print(x // y)
#2
print(x / y)
#2.2
print(x % y) 
#2
print(x ** y)

print(c)
```

```python
#uner

friends = 0

# Augmented Assignment
friends += 1  # Sama dengan: friends = friends + 1
print(f"Ditambah 1: {friends}")

friends += 2  # Penambahan
print(f"Ditambah 2: {friends}")

friends *= 3  # Perkalian
print(f"Dikali 3: {friends}")

friends /= 2  # Pembagian (hasilnya selalu float)
print(f"Dibagi 2: {friends}")

friends **= 2 # Eksponen (pangkat)
print(f"Dipangkat 2: {friends}")

# Modulus (Sisa Bagi)
remainder = 10 % 3 
print(f"Sisa bagi 10 dibagi 3 adalah: {remainder}")
```

```python
a = 10
b = 3

c = a ** b

print(f"Hasil perkalian {a} * {b} = {c}")
```

### B. Fungsi Matematika Bawaan (Built-in)
Python memiliki beberapa fungsi matematika yang bisa langsung digunakan tanpa perlu mengimpor apa pun, seperti pembulatan atau mencari nilai tertinggi/terendah.

### C. Modul `math`
Untuk perhitungan matematika tingkat lanjut (seperti akar kuadrat, nilai Pi, atau pembulatan paksa ke atas/bawah), kita harus memanggil modul bawaan Python dengan cara menuliskan `import math` di awal.

```python
# 1. Fungsi Matematika Bawaan
print("--- Fungsi Built-in ---")
print("Round (Pembulatan 3.14):", round(3.14))
print("Absolut (Jarak dari 0 untuk -4):", abs(-4))
print("Pow (4 pangkat 3):", pow(4, 3))
print("Max (Nilai tertinggi):", max(3.14, 4, 5))
print("Min (Nilai terendah):", min(3.14, 4, 5))

# 2. Menggunakan Modul Math
print("\n--- Modul Math ---")
import math

print("Konstanta PI:", math.pi)       
print("Konstanta E:", math.e)        
print("Akar kuadrat dari 9:", math.sqrt(9))  
print("Pembulatan paksa ke atas (9.1):", math.ceil(9.1)) 
print("Pembulatan paksa ke bawah (9.9):", math.floor(9.9))
```
