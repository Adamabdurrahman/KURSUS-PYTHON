## 2. Row, Column, Cell

Dari tabel `buku` di atas, ada 3 istilah dasar yang wajib kamu pahami:

| Istilah | Bahasa Indonesia | Artinya |
|---|---|---|
| **Row** | Baris | Satu catatan/record data yang lengkap (contoh: satu baris = data 1 buku) |
| **Column** | Kolom | Satu jenis atribut/informasi yang dimiliki setiap baris (contoh: `judul`, `penulis`) |
| **Cell** | Sel | Satu nilai spesifik, pertemuan antara 1 baris dan 1 kolom |

Contoh konkret dari tabel `buku` di atas:
- **Row ke-2** = seluruh data tentang buku "Bumi Manusia" (id 2, penulis Pramoedya Ananta Toer, terbit 1980)
- **Column `penulis`** = kumpulan semua nama penulis di tabel itu (Andrea Hirata, Pramoedya Ananta Toer, Ahmad Fuadi)
- **Cell** di row ke-3, column `tahun_terbit` = nilai `2009`

## 3. Tipe Data di PostgreSQL

Setiap column di database harus punya **tipe data** тАФ supaya database tahu jenis nilai apa yang boleh disimpan di situ. Ini beberapa tipe data dasar yang paling sering dipakai:

| Tipe Data | Untuk menyimpan | Contoh nilai |
|---|---|---|
| `INTEGER` (`INT`) | Bilangan bulat | `1`, `2005`, `-7` |
| `NUMERIC(p, s)` | Bilangan desimal presisi tetap тАФ cocok untuk uang/harga | `45000.50` |
| `REAL` / `DOUBLE PRECISION` | Bilangan desimal floating-point тАФ cocok untuk data sains | `3.14159` |
| `VARCHAR(n)` | Teks dengan panjang maksimum n karakter | `'Laskar Pelangi'` |
| `TEXT` | Teks tanpa batas panjang | Isi sebuah artikel panjang |
| `BOOLEAN` | Benar/salah | `TRUE`, `FALSE` |
| `DATE` | Tanggal saja | `'2005-08-01'` |
| `TIMESTAMP` | Tanggal + jam | `'2005-08-01 14:30:00'` |
| `SERIAL` | Integer yang otomatis bertambah (mirip `AUTO_INCREMENT` di MySQL) | dipakai untuk id/primary key |

> ЁЯТб **Kenapa `NUMERIC` dan bukan `REAL` untuk harga?** `REAL`/`DOUBLE PRECISION` itu floating-point тАФ bisa punya pembulatan kecil yang tidak presisi. Untuk uang, kita mau angka yang presisi persis, jadi pakai `NUMERIC(10,2)` (artinya: total 10 digit, 2 di antaranya di belakang koma).

## 6. `CREATE TABLE` тАФ Membuat Tabel

Sekarang kita buat tabel `buku` sungguhan di `db_perpustakaan`, sesuai dengan konsep row/column/cell dan tipe data yang baru kita pelajari.

```sql
CREATE TABLE buku (
    id_buku       SERIAL PRIMARY KEY,
    judul         VARCHAR(150) NOT NULL,
    penulis       VARCHAR(100),
    tahun_terbit  INTEGER,
    harga         NUMERIC(10,2),
    stok          INTEGER DEFAULT 0,
    tersedia      BOOLEAN DEFAULT TRUE
);
```

Penjelasan tiap baris:
- `id_buku SERIAL PRIMARY KEY` тЖТ id otomatis bertambah, jadi identitas unik tiap buku
- `judul VARCHAR(150) NOT NULL` тЖТ wajib diisi, maksimal 150 karakter
- `stok INTEGER DEFAULT 0` тЖТ kalau tidak diisi saat insert, otomatis bernilai 0
- `tersedia BOOLEAN DEFAULT TRUE` тЖТ kalau tidak diisi, otomatis dianggap tersedia

Jalankan di cell berikut:


````python
%%sql
CREATE TABLE buku (
    id_buku       SERIAL PRIMARY KEY,
    judul         VARCHAR(150) NOT NULL,
    penulis       VARCHAR(100),
    tahun_terbit  INTEGER,
    harga         NUMERIC(10,2),
    stok          INTEGER DEFAULT 0,
    tersedia      BOOLEAN DEFAULT TRUE
);
````

### `CREATE TABLE IF NOT EXISTS`

Kalau kamu tidak yakin apakah tabelnya sudah pernah dibuat sebelumnya, tambahkan `IF NOT EXISTS` supaya tidak error saat dijalankan ulang:

```sql
CREATE TABLE IF NOT EXISTS buku (
    id_buku SERIAL PRIMARY KEY,
    judul VARCHAR(150)
);
```


## 7. Melihat Struktur Tabel

Di MySQL/MariaDB (yang mungkin sudah pernah kamu pakai) ada perintah `DESCRIBE nama_tabel;`. Di PostgreSQL:
- Kalau pakai terminal `psql`, perintahnya adalah `\d buku`
- Kalau ingin lewat query SQL biasa (bisa dijalankan dari notebook ini), pakai `information_schema`:


````python
%%sql
SELECT column_name, data_type, is_nullable, column_default
FROM information_schema.columns
WHERE table_name = 'buku';
````

## 8. `ALTER TABLE` тАФ Mengubah Struktur Tabel

Setelah tabel dibuat, kita sering perlu mengubahnya: menambah column, mengganti tipe data, mengganti nama, atau menghapus column. Ini beberapa contoh:

### a) Menambah column baru тАФ `ADD COLUMN`


````python
%%sql
ALTER TABLE buku
ADD COLUMN penerbit VARCHAR(100);
````

### b) Mengubah tipe data column тАФ `ALTER COLUMN ... TYPE`

> ЁЯТб Di MySQL kamu mungkin terbiasa pakai `CHANGE`. Di PostgreSQL syntax-nya beda, pakai `ALTER COLUMN ... TYPE`.


````python
%%sql
ALTER TABLE buku
ALTER COLUMN penerbit TYPE VARCHAR(150);
````

### c) Mengganti nama column тАФ `RENAME COLUMN`

````python
%%sql
ALTER TABLE buku
RENAME COLUMN penerbit TO nama_penerbit;
````

### d) Menghapus column тАФ `DROP COLUMN`

````python
%%sql
ALTER TABLE buku
DROP COLUMN nama_penerbit;
````

### e) Mengganti nama tabel тАФ `RENAME TO`

Sekadar demonstrasi, kita ganti nama tabelnya lalu kembalikan lagi ke `buku` supaya modul berikutnya tetap konsisten.


````python
%%sql
ALTER TABLE buku RENAME TO data_buku;
````

````python
%%sql
ALTER TABLE data_buku RENAME TO buku;
````

Cek lagi strukturnya untuk memastikan tabel `buku` sudah kembali seperti semula (kolom `penerbit`/`nama_penerbit` sudah tidak ada lagi karena sempat kita hapus):


````python
%%sql
SELECT column_name, data_type
FROM information_schema.columns
WHERE table_name = 'buku';
````
