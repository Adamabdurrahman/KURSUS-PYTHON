## 1. Teori: Pernyataan `if`, `elif`, dan `else`

**Percabangan** memungkinkan program untuk menjalankan blok kode tertentu hanya jika sebuah kondisi bernilai `True`.

Strukturnya:
```python
if kondisi_1:
    # kode dijalankan jika kondisi_1 True
elif kondisi_2:
    # kode dijalankan jika kondisi_1 False, tapi kondisi_2 True
else:
    # kode dijalankan jika semua kondisi di atas False
```

> **Penting:** Python menggunakan **indentasi** (tab/spasi) untuk menandai blok kode. Blok kode di dalam `if` harus menjorok ke dalam!

````python
# Contoh 1: Cek kelayakan usia
# Perhatikan urutan pengecekan kondisi dari atas ke bawah!
age = int(input("Masukkan umurmu: "))

if age >= 100:
    print("Kamu terlalu tua untuk mendaftar")
elif age >= 18:
    print("Pendaftaran berhasil!")
elif age < 0:
    print("Kamu belum lahir!")
else:
    print("Kamu harus berusia 18+ untuk mendaftar")
````

````python
# Contoh 2: Cek input teks dari user
response = input("Apakah kamu mau makanan? (Y/T): ")

if response == "Y":
    print("Ini makanannya!")
else:
    print("Oke, tidak ada makanan untukmu!")
````

````python
# Contoh 3: Mengecek apakah input kosong
# String kosong ("") dianggap False oleh Python
name = input("Masukkan namamu: ")

if name == "":
    print("Kamu tidak memasukkan nama!")
else:
    print(f"Halo, {name}!")
````

````python
# Contoh 4: Menggunakan variabel Boolean sebagai kondisi
# Variabel bertipe bool bisa langsung dipakai sebagai kondisi tanpa perlu == True
online = False

if online:
    print("Kamu sedang online")
else:
    print("Kamu sedang offline")
````

## 2. Teori: Logical Operators

**Logical Operators** digunakan untuk menggabungkan dua atau lebih kondisi dalam satu pernyataan `if`.

| Operator | Fungsi | Contoh |
|----------|--------|--------|
| `and` | `True` jika **semua** kondisi bernilai `True` | `age > 18 and has_id == True` |
| `or` | `True` jika **minimal satu** kondisi bernilai `True` | `temp < 0 or temp > 40` |
| `not` | Membalik nilai boolean (`True` → `False`, begitu pula sebaliknya) | `not is_banned` |

````python
# Contoh penggunaan 'or': kondisi cuaca
temp = 20
sunny = True

# 'or': cukup satu kondisi yang True agar blok ini dijalankan
if temp <= 0 or temp >= 30:
    print("Cuaca tidak nyaman")
else:
    print("Cuaca nyaman!")

# 'not': membalik nilai boolean
if not sunny:
    print("Cuaca mendung")
else:
    print("Cuaca cerah")
````

````python
# Contoh penggunaan 'and': akses login
username = "admin"
password = "123456"

# 'and': SEMUA kondisi harus True
if username == "admin" and password == "123456":
    print("Login berhasil!")
else:
    print("Username atau password salah!")
````

````python
# Contoh penggunaan 'and': akses login
username = "admin"
password = "123456"

# 'and': SEMUA kondisi harus True
if username == "admin":
    if password == "123456":
        print("Login berhasil!")
    else:
        print("Password salah!")
else:
    print("Username salah!")
````

## 3. Teori: Conditional Expression (Ternary Operator)

**Conditional expression** adalah cara ringkas untuk menulis `if-else` sederhana dalam **satu baris**.

```python
# Bentuk panjang
if kondisi:
    nilai = "A"
else:
    nilai = "B"

# Bentuk ternary (satu baris)
nilai = "A" if kondisi else "B"
```

> **Kapan dipakai?** Gunakan ternary untuk kondisi yang **sederhana**. Jika kondisinya kompleks, lebih baik tetap gunakan `if-else` biasa agar kode tetap mudah dibaca.

````python
if true:
    print("A")
else:
    print("B")
    
print("A") if true else print("B") 
````

## 4. Teori: String Methods

**String methods** adalah fungsi-fungsi bawaan yang bisa dipanggil langsung pada sebuah variabel string menggunakan notasi titik (`.`). Method ini tidak mengubah string aslinya, melainkan **mengembalikan string baru**.

| Method | Fungsi |
|--------|--------|
| `len(teks)` | Menghitung panjang/jumlah karakter |
| `.find(" ")` | Mencari posisi/index karakter pertama yang ditemukan. Mengembalikan `-1` jika tidak ada |
| `.capitalize()` | Huruf pertama jadi kapital, sisanya kecil |
| `.upper()` | Semua karakter menjadi huruf besar |
| `.lower()` | Semua karakter menjadi huruf kecil | 
| `.isdigit()` | `True` jika semua karakter adalah angka |
| `.isalpha()` | `True` jika semua karakter adalah huruf |
| `.count("x")` | Menghitung berapa kali karakter/substring muncul |
| `.replace("a", "b")` | Mengganti semua kemunculan "a" dengan "b" |

````python
# Demo semua string methods
name = "budi santoso"
phone_number = "081-234-567-890"

print("--- String Methods pada Nama ---")
print(f"Teks asli          : '{name}'")
print(f"Panjang karakter   : {len(name)}")
print(f"Posisi spasi (' ') : {name.find(' ')}")
print(f"capitalize()       : {name.capitalize()}")
print(f"upper()            : {name.upper()}")
print(f"lower()            : {name.lower()}")
print(f"isdigit()          : {name.isdigit()}")  # False, karena ada huruf
print(f"isalpha()          : {name.isalpha()}")  # False, karena ada spasi
print(f"isspace()          : {name.isspace()}")  # False, karena ada huruf
print(f"replace(' ', '_') : {name.replace(' ', '_')}")  # Mengganti spasi dengan underscore

print("\n--- String Methods pada Nomor HP ---")
print(f"Teks asli          : '{phone_number}'")
print(f"Jumlah tanda '-'   : {phone_number.count('-')}")
clean_phone = phone_number.replace("-", "")
print(f"Setelah replace('-', ''): '{clean_phone}'")
````

Sekarang saatnya kita gabungkan semua materi! Kita akan membuat tiga program interaktif.

````python
# EXERCISE 1: Kalkulator Python
# Menggabungkan: input, type casting, if/elif/else, dan operasi matematika
print("--- Kalkulator Python ---")

operator = input("Masukkan operator (+ - * / ** % //): ")
num1 = float(input("Masukkan angka pertama: "))
num2 = float(input("Masukkan angka kedua: "))

if operator == "+":
    result = num1 + num2
    print(round(result, 3))
elif operator == "-":
    result = num1 - num2
    print(round(result, 3))
elif operator == "*":
    result = num1 * num2
    print(round(result, 3))
elif operator == "/":
    result = num1 / num2
    print(round(result, 3))
elif operator == "**":
    result = num1 ** num2
    print(round(result, 3))
elif operator == "%":
    result = num1 % num2
    print(round(result, 3))
elif operator == "//":
    result = num1 // num2
    print(int(result))
else:
    print(f"'{operator}' bukan operator yang valid!")
````

````python
# EXERCISE 2: Konverter Berat & Suhu
# Menggabungkan: input, type casting, if/elif/else, dan operasi matematika
print("--- Konverter Berat ---")
weight = float(input("Masukkan beratmu: "))
unit = input("Kilogram atau Pound? (K/P): ")

if unit == "K":
    weight = weight * 2.205
    unit = "Lbs."
    print(f"Beratmu adalah: {round(weight, 1)} {unit}")
elif unit == "P":
    weight = weight / 2.205
    unit = "Kgs."
    print(f"Beratmu adalah: {round(weight, 1)} {unit}")
else:
    print(f"'{unit}' bukan satuan yang valid!")

print("\n--- Konverter Suhu ---")
unit = input("Celsius atau Fahrenheit? (C/F): ")
temp = float(input("Masukkan suhu: "))

if unit == "C":
    temp = round((9 * temp) / 5 + 32, 1)
    print(f"Suhu dalam Fahrenheit: {temp}°F")
elif unit == "F":
    temp = round((temp - 32) * 5 / 9, 1)
    print(f"Suhu dalam Celsius: {temp}°C")
else:
    print(f"'{unit}' bukan satuan yang valid!")
````

````python
# EXERCISE 3: Validator Username
# Menggabungkan: input, string methods, if/elif/else, dan logical operator 'not'
print("--- Validator Username ---")
username = input("Masukkan username: ")

# Aturan:
# 1. Maksimal 12 karakter
# 2. Tidak boleh mengandung spasi
# 3. Hanya boleh huruf (tidak boleh ada angka/simbol)
if len(username) > 12:
    print("Username tidak boleh lebih dari 12 karakter!")
elif not username.find(" ") == -1:  # find() mengembalikan -1 jika tidak ditemukan
    # not false = true
    print("Username tidak boleh mengandung spasi!")
elif not username.isalpha():
    print("Username tidak boleh mengandung angka atau simbol!")
else:
    print(f"Selamat datang, {username}! Akun berhasil dibuat.")
````
